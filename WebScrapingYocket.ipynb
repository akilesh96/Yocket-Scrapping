```python
# Main code block for scrapping
class YocketScrapper:
    def __init__(self, url="https://yocket.com/universities?c_name=Canada&s_name=rank&s_order=asc&type=Public&level=2&pc_name=Computer%20Science"):
        self.output = []
        s=Service(ChromeDriverManager().install())
        options = webdriver.ChromeOptions()
        options.add_argument("user-data-dir=D:\\Python Projects\\Web Scraping\\Profile")
        self.driver = webdriver.Chrome(service=s, options=options)
        self.driver.maximize_window()
        self.driver.get(url)
                 
    def click_button(self):
        self.driver.find_element(by=By.XPATH, value='//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/div/div[2]/div[2]/button').click()
                 
    def count_universities(self):
        universities_count_xpath = '//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/div/div[1]/div/div[1]/div/p'
        universities_count_text = self.driver.find_element(by=By.XPATH, value=universities_count_xpath).text
        self.universities_count = int(re.search(r'\d+', universities_count_text[::-1]).group()[::-1])
        print("Universities Count:", self.universities_count)
                 
    def scrap_university(self):
        xpath_university_course = '//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/div/div[2]/div[1]/div[{}]/div/div/div[4]/div[2]/div[2]/p'
        university_name_xpath = '//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/div/div[2]/div[1]/div[{}]/div/div/div[2]/div[1]/a/p'
        course_xpath = '//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/div/div[2]/div[1]/div[{}]/div/div/div[4]/div[2]/div[2]/p'
        location_xpath = '//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/div/div[2]/div[1]/div[{}]/div/div/div[2]/div[2]/p'
        xpath_university_link = '//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/div/div[2]/div[1]/div[{}]/div/div/div[2]/div[1]/a'
        student_details_link = 'https://yocket.com/connect?university={}&status=7&course=Computer%20Science'
        self.universities_code = []
                 
        for university_id in range(1,self.universities_count+1):
            self.driver.execute_script("window.scrollTo(0, 0);")
            time.sleep(1)
#             university_path = xpath_university_course.format(university_id)

#             self.university_name = self.driver.find_element(by=By.XPATH, value=university_name_xpath.format(university_id)).text.strip()
#             self.course_name = self.driver.find_element(by=By.XPATH, value=course_xpath.format(university_id)).text.strip()
#             self.location = self.driver.find_element(by=By.XPATH, value=location_xpath.format(university_id)).text.strip()
            self.university_code = self.driver.find_element(by=By.XPATH, value=xpath_university_link.format(university_id)).get_attribute('href').strip().split("-")[-1]
            self.universities_code.append(self.university_code)
#             self.driver.execute_script("window.open('');")
#             self.driver.switch_to.window(self.driver.window_handles[1])
#             self.driver.get(student_details_link.format(self.university_code))
#             self.driver.find_element(by=By.XPATH, value=university_path).click()
#             self.student_details()
#             print(self.output)
        self.universities_code = list(set(self.universities_code))
                 
    def student_details(self):
        try:
#             self.driver.switch_to.window(self.driver.window_handles[1])
#             student_link_xpath = '//*[@id="Students"]/div[1]/div[2]/a'
            self.scroll_end()
            self.output.append({"university_code": self.university_code, "university": self.university_name, "course": self.course_name, "location": self.location})
            self.driver.close()
            self.driver.switch_to.window(self.driver.window_handles[0])
        except Exception as e:
            print(e)
            self.output.append({"university": self.university_name, "course": self.course_name, "location": self.location})
            self.driver.close()
            self.driver.switch_to.window(self.driver.window_handles[0])
            
    def scroll_end(self):
        SCROLL_PAUSE_TIME = 1

        # Get scroll height
        last_height = self.driver.execute_script("return document.body.scrollHeight")

        while True:
            # Scroll down to bottom
            self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait to load page
            time.sleep(SCROLL_PAUSE_TIME)

            # Calculate new scroll height and compare with last scroll height
            new_height = self.driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
```

```python
# Main code block for scrapping
xpath_student = '/html/body/div[1]/div/div/div[2]/div/main/div/div[3]/div[{}]/div[1]/div[2]/a'
student_name_xpath = '/html/body/div[1]/div/div/div[2]/div/main/div/div[3]/div[{}]/div[1]/div[2]/a/span'
intake_xpath = '/html/body/div[1]/div/div/div[2]/div/main/div/div[3]/div[{}]/div[1]/div[3]'
university_name_xpath = '/html/body/div[1]/div/div/div[2]/div/main/div/div[3]/div[{}]/div[2]/div[2]/p[1]/a'
course_xpath = '/html/body/div[1]/div/div/div[2]/div/main/div/div[3]/div[{}]/div[2]/div[2]/p[2]/span/a'
check_ielts_xpath = '/html/body/div[1]/div/div/div[2]/div/main/div/div[3]/div[{}]/div[3]/div/div[1]/p/span[1]'
ielts_score_xpath = '/html/body/div[1]/div/div/div[2]/div/main/div/div[3]/div[{}]/div[3]/div/div[1]/p/span[2]'
ug_course_xpath = '//*[@id="CollegeDetails"]/div[2]/div/div/div[2]/p[2]'
ug_school_xpath = '//*[@id="CollegeDetails"]/div[2]/div/div/div[2]/p[1]'
ug_grade_type_xpath = '//*[@id="CollegeDetails"]/div[2]/div/div/div[1]/p[2]'
ug_grade_score_xpath = '//*[@id="CollegeDetails"]/div[2]/div/div/div[1]/p[1]'

def handle_missing_data(u_id, xpath):
    try:
        value = ob.driver.find_element(by=By.XPATH, value=xpath.format(u_id)).text.strip()
    except Exception as e:
        value = ""
    return value

def handle_missing_data2(xpath):
    try:
        value = ob.driver.find_element(by=By.XPATH, value=xpath).text.strip()
    except Exception as e:
#         print(e)
        value = ""
    return value

student_log = open("Scrapped_Data\\student_log.txt", "a")

for university_code in ob.universities_code[23:]:
    university_student_history_url = 'https://yocket.com/connect?university={}&status=7&course=Computer%20Science'.format(university_code)

    ob.driver.get(university_student_history_url)
    student_count_xpath = '//*[@id="__layout"]/div/div[2]/div/main/div/div[2]/p[2]'
    time.sleep(5)
    try:
        student_count_text = ob.driver.find_element(by=By.XPATH, value=student_count_xpath).text
    except Exception as e:
        print("No Aspirants:", university_code)
        continue
    student_count = int(re.search(r'\d+', student_count_text[::-1]).group()[::-1])
    show_count = int(re.search(r'\d+', student_count_text).group())
    while show_count != student_count:
        ob.scroll_end()
        student_count_text = ob.driver.find_element(by=By.XPATH, value=student_count_xpath).text
        show_count = int(re.search(r'\d+', student_count_text).group())
        print(show_count, student_count)
        
    student_details_list = []

    for student_id in range(1,student_count+1):
        ob.driver.execute_script("window.scrollTo(0, 0);")
        time.sleep(1)

        student_name = handle_missing_data(student_id, student_name_xpath)
        course_name = handle_missing_data(student_id, course_xpath)
        intake = handle_missing_data(student_id, intake_xpath).replace("\n", " ")
        university_name = handle_missing_data(student_id, university_name_xpath)
        check_ielts = handle_missing_data(student_id, check_ielts_xpath)
        ielts_score = None
        if check_ielts == "IELTS":
            ielts_score = handle_missing_data(student_id, ielts_score_xpath)

        try:
            ob.driver.find_element(by=By.XPATH, value=xpath_student.format(student_id)).click()
        except Exception as e:
            print("Student ID:", student_id)
            continue
        ob.driver.switch_to.window(ob.driver.window_handles[1])
        time.sleep(10)
        ug_course = handle_missing_data2(ug_course_xpath)
        ug_school = handle_missing_data2(ug_school_xpath)
        ug_grade_type = handle_missing_data2(ug_grade_type_xpath)
        ug_grade_score = handle_missing_data2(ug_grade_score_xpath)

        print(student_id, student_name, course_name, intake, university_name, ielts_score, ug_course, ug_grade_type, ug_grade_score, ug_school)
        student_details_list.append([student_id, student_name, course_name, intake, university_name, ielts_score, ug_course, ug_grade_type, ug_grade_score, ug_school])
        ob.driver.close()
        ob.driver.switch_to.window(ob.driver.window_handles[0])
        
    df = pd.DataFrame(student_details_list, columns = ["ID", "Student Name", "Masters", "Intake", "University Name", "IELTS", "UG Degree", "UG Grade Type", "UG Grade Score", "UG School"])
    df.to_csv("Scrapped_Data\\"+university_name.replace(" ","_")+".csv", index=False)
    student_log.write(university_name+": "+str(student_count)+"\n")
    print(university_name+": "+str(student_count))
student_log.close()
```
